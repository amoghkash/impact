```python
import matplotlib.pyplot as plt
import csv
```

```python
xx = []
xy = []
xz = []
mag = []
y=[]
```

```python
with open('data0.csv') as file:
    next(file)
    x =100
    lines = csv.reader(file, delimiter=',')
    for row in lines:
        if x==100:
            xx.append(float(row[0]))
            xy.append(float(row[1]))
            xz.append(float(row[2]))
            mag.append(float(row[3]))
            time = float(row[4])
            time = round(time, 2)
            y.append(time)
            x=0
        x = x +1
```

```python
print(mag)
print(y)
```

```python
plt.plot(y, mag, label = "Magnitude")
plt.plot(y, xz, label = "Z Acc")
```

```python
plt.legend()
plt.show()
```
